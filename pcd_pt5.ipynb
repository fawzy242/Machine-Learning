```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import urllib.request
import os
# Download gambar sample
filename = "test-logo.jpg"
# if not os.path.exists(filename):
#  urllib.request.urlretrieve(url, filename)
# Membaca citra dalam mode grayscale
image = cv2.imread(filename, 0)
# Menentukan konstanta skala (c)
c = 255 / np.log(1 + np.max(image))
# Menerapkan transformasi logaritma pada citra
log_transform = c * (np.log(1 + image.astype(np.float64)))
# Normalisasi hasil transformasi agar sesuai dengan format citra 8-bit
log_transform = np.array(log_transform, dtype=np.uint8)
# Menampilkan citra asli dan hasil transformasi logaritma
plt.figure(figsize=(10, 5))
# Menampilkan citra asli
plt.subplot(1, 2, 1)
plt.title('Citra Asli')
plt.imshow(image, cmap='gray')
plt.axis('off') # Menghilangkan axis
# Menampilkan hasil transformasi logaritma
plt.subplot(1, 2, 2)
plt.title('Transformasi Logaritma')
plt.imshow(log_transform, cmap='gray')
plt.axis('off') # Menghilangkan axis
# Menampilkan plot
plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import urllib.request
import os
# Download gambar sample
url = "https://raw.githubusercontent.com/opencv/opencv/master/samples/data/lena.jpg"
filename = "gw.jpg"
if not os.path.exists(filename):
 urllib.request.urlretrieve(url, filename)
# Membaca citra dalam mode grayscale
image = cv2.imread(filename, 0)
# Fungsi untuk melakukan Gamma Correction
def gamma_correction(image, gamma):
 # Konversi citra ke tipe float32 untuk menghindari clipping
 image_float = image.astype(np.float32) / 255.0

 # Melakukan transformasi pangkat-law (gamma correction)
 corrected_image = np.power(image_float, gamma)
 # Mengembalikan ke rentang 0-255 dan konversi ke tipe uint8
 corrected_image = np.uint8(corrected_image * 255)
 return corrected_image
# Nilai gamma
gamma = 2.2 # Ubah nilai gamma untuk melihat perbedaan
corrected_image = gamma_correction(image, gamma)
# Tampilkan citra asli dan citra setelah gamma correction
plt.figure(figsize=(10, 5))
# Citra Asli
plt.subplot(1, 2, 1)
plt.title('Citra Asli')
plt.imshow(image, cmap='gray')
plt.axis('off')
# Citra setelah Gamma Correction
plt.subplot(1, 2, 2)
plt.title('Citra Setelah Gamma Correction (gamma = {})'.format(gamma))
plt.imshow(corrected_image, cmap='gray')
plt.axis('off')
plt.tight_layout()
plt.show()
```
