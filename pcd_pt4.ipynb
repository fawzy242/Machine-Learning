```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

print("OpenCV version:", cv2.__version__)
```

```python
# Ganti nama file sesuai gambar kamu
img_path = "test-logo.jpg"

# Baca gambar dalam mode grayscale
gambar = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

if gambar is None:
    raise FileNotFoundError(f"Gambar '{img_path}' tidak ditemukan.")

plt.imshow(gambar, cmap='gray')
plt.title('Gambar Asli')
plt.axis('off')
plt.show()
```

```python
hist_asli = cv2.calcHist([gambar], [0], None, [256], [0, 256])

plt.figure()
plt.title('Histogram Gambar Asli')
plt.xlabel('Intensitas Pixel')
plt.ylabel('Jumlah Pixel')
plt.plot(hist_asli)
plt.xlim([0, 256])
plt.grid(True)
plt.show()
```

```python
gambar_ekual = cv2.equalizeHist(gambar)

plt.imshow(gambar_ekual, cmap='gray')
plt.title('Gambar Setelah Histogram Equalization')
plt.axis('off')
plt.show()
```

```python
hist_eq = cv2.calcHist([gambar_ekual], [0], None, [256], [0, 256])

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.title('Histogram Asli')
plt.plot(hist_asli)
plt.xlim([0,256])
plt.grid(True)

plt.subplot(1,2,2)
plt.title('Histogram Ekualisasi')
plt.plot(hist_eq)
plt.xlim([0,256])
plt.grid(True)

plt.tight_layout()
plt.show()
```
